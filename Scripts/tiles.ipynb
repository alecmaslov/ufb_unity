```python
from PIL import Image
import os
import matplotlib.pyplot as plt
```

```python
def display(image, figsize=(8,8)):
    plt.figure(figsize=figsize)
    plt.axis('off')
    plt.imshow(image)
    plt.show()
```

```python
tilemap_file = "E:/UnityProjects/UFB/Assets/Resources/Images/Map__KrakenCropped.png"
```

```python
image_size = (5642, 5642) # this divides by 26 to 217
image = Image.open(tilemap_file)
image = image.resize(image_size)
```

```python
display(image)
```

```python
map_name = "Kraken"
outpath = f"E:/UnityProjects/UFB/Assets/Resources/Maps/{map_name}"

if not os.path.exists(outpath):
    os.makedirs(outpath)
```

```python
image.size
```

```python
num_tiles = 26
horizontal_tile_names = [x for x in "ABCDEFGHIJKLMNOPQRSTUVWXYZ"]
output_size = (256, 256)

image_width, image_height = image.size
tile_width = image_width // num_tiles
tile_height = image_height // num_tiles

print(f"Tile size: {tile_width} x {tile_height}")

for i in range(num_tiles):
    for j in range(num_tiles):
        tile = image.crop((i*tile_width, j*tile_height, (i+1)*tile_width, (j+1)*tile_height))

        row_name = j
        col_name = horizontal_tile_names[i]

        outfile = os.path.join(outpath, f"Tile__{i}_{j}.png")
        tile = tile.resize(output_size)
        tile.save(f"E:/UnityProjects/UFB/Assets/Resources/Maps/tile_{col_name}_{row_name}.png")
```
